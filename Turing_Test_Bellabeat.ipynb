#### **a.) Relevance**

From similar data on Kaggle; Cardiovascular Disease Dataset - https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset, we are able to derive the following units of measurement:

* Age - days

* Gender - 1:Female, 2:Male

* Height - cm

* Weight - kg

* Cholesterol - 1: normal, 2: above normal, 3: well above normal

* Smoke - 1:Active smoker, 0:Doesn't smoke

* Alco - 1: Alcohol intake 0: No alcohol



```python
# convert age to years

cardio['age'] = cardio['age'] / 365

cardio['age'] = cardio['age'].apply(lambda x: math.floor(x))
```

```python
# age groups

print(f"Min Age: {cardio['age'].min()} , Max Age: {cardio['age'].max()}")

cardio['age_group'] = '00'

cardio.loc[cardio['age'] < 20, 'age_group'] = 'Teen'

cardio.loc[(cardio['age'] >= 20) & (cardio['age'] <= 35), 'age_group'] = 'Youth'

cardio.loc[(cardio['age'] > 35) & (cardio['age'] <= 60), 'age_group'] = 'Adult'

cardio.loc[(cardio['age'] > 60), 'age_group'] = 'Senior'

print(f"Age groups: {cardio['age_group'].unique()}")
```

```python
# calculate bmi [ weight in kg / (height im metres)^2 ]

cardio['bmi'] = cardio["weight"] / (cardio["height"]/100)**2
```

```python
# replace gender values

cardio['gender'] = np.where(cardio['gender']==1, 'Female', 'Male')

cardio['gender'].unique()
```

```python
# Replace cholesterol values

cardio.loc[cardio['cholesterol'] == 1, 'cholesterol'] = 'Normal'

cardio.loc[cardio['cholesterol'] == 2, 'cholesterol'] = 'Above Normal'

cardio.loc[cardio['cholesterol'] == 3, 'cholesterol'] = 'Well above Normal'

cardio['cholesterol'].unique()
```
